Sonifying the data by the type of device

```python
import pandas as pd
from midiutil import MIDIFile
import math
```

```python
bluetooth=pd.read_csv('canleybrook/bluetooth.txt', header=None)
company_db = pd.read_csv('companies.csv', header=None)
company_db.head()
```

```python
start = bluetooth[0].min()
start
```

```python
def clean_organisation(org):
    o = org.replace('{', '').replace('}','').split('=')[0].strip()
    if o == '':
        return -1
    
    return o

bluetooth['company'] = bluetooth[7].apply(clean_organisation)
bluetooth
```

```python
companys = {}
for c in company_db.to_numpy():
    companys[c[0]] = c[3]
companys
```

```python
bluetooth['type'] = bluetooth['company'].map(companys)
bluetooth.head(10)
```

```python
import json
orgs = []
for blue in bluetooth.to_numpy():
    org = ''
    if blue[10] != -1:
        org = companys[int(blue[10])]
    else:
        org = 'unknown'
    orgs.append({'time': int(blue[0]), 'company': int(blue[10]), 'type': org, 'signal':int(blue[2])}) 
print(json.dumps({'blue': orgs}))
```
