```python
import pandas as pd
```

```python
def clean_organisation(org):
    return org.replace('{', '').replace('}','').split('=')[0]
```

```python
loc = pd.read_csv('canleybrook/bluetooth.txt', header=None)
loc['company'] = loc[7].apply(clean_organisation)
loc
```

```python
loc1 = pd.read_csv('stoney_stanton/bluetooth.txt', header=None)
loc1['company'] = loc1[7].apply(clean_organisation)
loc1
```

```python
loc2 = pd.read_csv('fargo/bluetooth.txt', header=None)
loc2['company'] = loc2[7].apply(clean_organisation)
loc2
```

```python
loc3 = pd.read_csv('crackley/bluetooth.txt', header=None)
loc3['company'] = loc3[7].apply(clean_organisation)
loc3
```

```python
loc4 = pd.read_csv('musicmile/bli', header=None)
loc4['company'] = loc4[7].apply(clean_organisation)
loc4
```

```python
locations = pd.concat([loc, loc1, loc2, loc3])
locations.dtypes
```

```python
coms = pd.DataFrame(locations['company'].unique())
coms.to_csv('companies.csv')
```
