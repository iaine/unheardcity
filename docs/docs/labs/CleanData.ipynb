```python
import pandas as pd
import hashlib
```

```python
filename="musicmile/bluetooth_1634469751686.txt"
raw_data = pd.read_csv(filename, header=None)
raw_data.columns=['Time', 'Id','RSSI', 'Primary', 'Secondary', 'AdInterval','DeviceName','ManufacturerData', 'TxPower', 'AdvertiseFlag']
raw_data.head()
```

```python
def hash_id(device_id):
    return hashlib.sha224(str.encode(device_id)).hexdigest()

raw_data['DeviceId'] = raw_data['Id'].apply(hash_id)
```

```python
del raw_data['Id']
raw_data.head()
```

```python
raw_data.to_csv('musicmile/hash_bluetooth.csv')
```
